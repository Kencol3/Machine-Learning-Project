```python
import  numpy as np
from scipy import stats
import matplotlib.rcsetup as rcsetup
```

```python
#get rid of outliers by computing abs value of zscore and removing points more than 3 sd from mean
lot_price = housing[["LotArea", "SalePrice"]]
lot_price = lot_price[(np.abs(stats.zscore(lot_price)) < 4).all(axis=1)]
plt.scatter(x = lot_price.LotArea, y = lot_price.SalePrice)
plt.show()
```

```python
#merge outlier dataset w/ original dataset for new dataset w/ outliers
y = pd.merge(housing2, lot_price, how = "inner", on = ["LotArea","SalePrice"])
y = y[-y.duplicated()]
```
